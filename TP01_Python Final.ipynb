## SVM Support Vector Machines

```python
#import necessary libraries
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LogisticRegression
from sklearn import svm
```

```python
#A
#Generate data

np.random.seed(1693)
X1 = np.random.uniform(0, 1, 500) - 0.5
X2 = np.random.uniform(0, 1, 500) - 0.5
Y = ((X1**2 - X2**2) > 0).astype(int)
```

```python
#B
#Plot data

color= ['red' if i == 0 else 'blue' for i in Y]

fig = plt.figure(figsize=(15, 8))
ax = fig.add_subplot(111)

plt.scatter(X1, X2, color=color)
plt.grid()
plt.show()
```

```python
#C
#Fit Logistic Regression

#Combine both X1 and X2 side by side
X = np.column_stack((X1,X2))

#Create a Logistic Regression Classifier (clf)
clf = LogisticRegression(random_state=1693, fit_intercept=True)
clf.fit(X, Y)
```

```python
#D

#Plot predicitions and 

Y_hat = clf.predict(X)

color= ['red' if i == 0 else 'blue' for i in Y_hat]

fig = plt.figure(figsize=(15, 8))
ax = fig.add_subplot(111)

plt.scatter(X1, X2, color=color)
plt.grid()
plt.show()
```

```python
#E

#Creating Non-Linear Transformations
X2_2 = X2**2
X1X2 = X1*X2

#Creating the new Array of Transformed X Arrays
X_NonLin = np.column_stack((X1, X2_2, X1X2))

#Create a Logistic Regression Classifier (clf)
clf = LogisticRegression(random_state=1693, fit_intercept=True)
clf.fit(X_NonLin, Y)
Y_hat = clf.predict(X_NonLin)

color= ['red' if i == 0 else 'blue' for i in Y_hat]

fig = plt.figure(figsize=(15, 8))
ax = fig.add_subplot(111)

plt.scatter(X1, X2, color=color)
plt.grid()
plt.show()
```

```python
#G

# Fit the SVM model with a Linear Kernel
clf = svm.SVC(kernel='linear', C=0.1)
clf.fit(X, Y)
Y_hat = clf.predict(X)

color= ['red' if i == 0 else 'blue' for i in Y_hat]

fig = plt.figure(figsize=(15, 8))
ax = fig.add_subplot(111)

plt.scatter(X1, X2, color=color)


ax.set_ylim(-0.5, 0.5)
ax.set_xlabel('X1')
ax.set_ylabel('X2')
ax.grid()
ax.set_title("Linear Kernel")
```

```python
#F

# Fit the SVM model with a Polynomial Kernel
clf = svm.SVC(kernel='poly', degree=2, C = 0.1)
  #OR
#clf = svm.SVC(kernel = "rbf", C = 0.1)

clf.fit(X, Y)
Y_hat = clf.predict(X)

color= ['red' if i == 0 else 'blue' for i in Y_hat]

fig = plt.figure(figsize=(15, 8))
ax = fig.add_subplot(111)

plt.scatter(X1, X2, color=color)


ax.set_ylim(-0.5, 0.5)
ax.set_xlabel('X1')
ax.set_ylabel('X2')
ax.grid()
ax.set_title("Polynomial Kernel (Degree: 2)")
```

```python
#I

#Comments:
'''
Logistic Regression and other techniques we have learned often can only do linear decision boundaries. That's why the logisic regression showed a straight line separating the data even though
the actual data was more complicated than that.

SVM is useful for non-linear decision boundaries, though it can also do linear boundaries if applicable. In our case we made a decision boundary that was not able to separate the two classes with 
one straight line. That is where non-linear kernels for SVM come in. 

We could change the non-linear kernel to get different outputs as well as tuning the cost or gamma parameters to change the boundaries if the problem called for that. Our data was fairly simple
and so basic cost and gamma values worked well to differentiate the two classes.

'''
```
